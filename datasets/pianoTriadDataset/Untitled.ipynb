```python
import os
from scipy.io.wavfile import read, write
from IPython.display import display, Audio
import numpy as np
import nlpaug
import nlpaug.augmenter.audio as naa
import librosa
import librosa.display as librosa_display
import matplotlib.pyplot as plt

path_original = './audio/piano_2_An_m_f_0.wav'

eps = 1e-10

notes_sel = ['Bn']
triads = ['d']
octaves = ['2','3']
volumes = ['f','m','p']
examples = ['00','01','02','03','04','05','06','07','08','09']



n = 0
for triad in triads:
    for octave in octaves:
        for volume in volumes:
            for example in examples:
                for note_sel in notes_sel:
                    n = n+1
                    try:
                        original_name = path_original[0:14]+octave+'_'+note_sel+'_'+triad+'_'+volume+'_'+example+'.wav'
                        print(' ')
                        print(n)
                        print('Original:')
                        print(original_name)
                        print(note_sel)
                        original = read(original_name)
                        original = original[1]/32768
                        original = original*0.9
                        control = 1
                    except:
                        control = 0
                        
                        
                    if control ==1:
                        aug = naa.PitchAug(sampling_rate=16000, zone=(0,1), factor=(-3,-2))
                        aug1 = aug.augment(original)
                        np.clip(aug1, 0+eps, 1-eps)
                        aug1 = aug1*32768
                        aug1 = aug1.astype(np.int16)
                        
                        aug = naa.PitchAug(sampling_rate=16000, zone=(0,1), factor=(-2,-1))
                        aug2 = aug.augment(original)
                        np.clip(aug2, 0+eps, 1-eps)
                        aug2 = aug2*32768
                        aug2 = aug2.astype(np.int16)

                        aug = naa.PitchAug(sampling_rate=16000, zone=(0,1), factor=(-1,0))
                        aug3 = aug.augment(original)
                        np.clip(aug3, 0+eps, 1-eps)
                        aug3 = aug3*32768
                        aug3 = aug3.astype(np.int16)

                        aug = naa.PitchAug(sampling_rate=16000, zone=(0,1), factor=(1,2))
                        aug4 = aug.augment(original)
                        np.clip(aug4, 0+eps, 1-eps)
                        aug4 = aug4*32768
                        aug4 = aug4.astype(np.int16)

                        aug = naa.PitchAug(sampling_rate=16000, zone=(0,1), factor=(2,3))
                        aug5 = aug.augment(original)
                        np.clip(aug5, 0+eps, 1-eps)
                        aug5 = aug5*32768
                        aug5 = aug5.astype(np.int16)
                        
                        aug = naa.PitchAug(sampling_rate=16000, zone=(0,1), factor=(3,4))
                        aug6 = aug.augment(original)
                        np.clip(aug6, 0+eps, 1-eps)
                        aug6 = aug6*32768
                        aug6 = aug6.astype(np.int16)

                        if note_sel == 'Bn':
                            if octave == '2':
                                new_name = path_original[0:14]+octave+'_Af_'+triad+'_'+volume+'_'+example+'.wav'
                                print(new_name)
                                write(new_name, 16000, aug1)
                                new_name = path_original[0:14]+octave+'_An_'+triad+'_'+volume+'_'+example+'.wav'
                                print(new_name)
                                write(new_name, 16000, aug2)
                                new_name = path_original[0:14]+octave+'_Bf_'+triad+'_'+volume+'_'+example+'.wav'
                                print(new_name)
                                write(new_name, 16000, aug3)
                                new_name = path_original[0:14]+'3_Cn_'+triad+'_'+volume+'_'+example+'.wav'
                                print(new_name)
                                write(new_name, 16000, aug4)
                                new_name = path_original[0:14]+'3_Df_'+triad+'_'+volume+'_'+example+'.wav'
                                print(new_name)
                                write(new_name, 16000, aug5)
                                new_name = path_original[0:14]+'3_Dn_'+triad+'_'+volume+'_'+example+'.wav'
                                print(new_name)
                                write(new_name, 16000, aug6)
                            elif octave == '3':
                                new_name = path_original[0:14]+octave+'_Af_'+triad+'_'+volume+'_'+example+'.wav'
                                print(new_name)
                                write(new_name, 16000, aug1)
                                new_name = path_original[0:14]+octave+'_An_'+triad+'_'+volume+'_'+example+'.wav'
                                print(new_name)
                                write(new_name, 16000, aug2)
                                new_name = path_original[0:14]+octave+'_Bf_'+triad+'_'+volume+'_'+example+'.wav'
                                print(new_name)
                                write(new_name, 16000, aug3)
                                new_name = path_original[0:14]+'4_Cn_'+triad+'_'+volume+'_'+example+'.wav'
                                print(new_name)
                                write(new_name, 16000, aug4)
                                new_name = path_original[0:14]+'4_Df_'+triad+'_'+volume+'_'+example+'.wav'
                                print(new_name)
                                write(new_name, 16000, aug5)
                                new_name = path_original[0:14]+'4_Dn_'+triad+'_'+volume+'_'+example+'.wav'
                                print(new_name)
                                write(new_name, 16000, aug6)
                                
                    

                        else:
                            print('FAIL')
```
